```python
input_tokens = tokenizer("Write a python code for adding two numbers", return_tensors="pt")
input_tokens
```

```python
import torch
from transformers import AutoModelForCausalLM

model = AutoModelForCausalLM.from_pretrained("bert-base-uncased", torch_dtype = torch.bfloat16)
```

```python
out = model(input_ids = input_tokens["input_ids"])
out
#predictions are made by the model
```

```python
gen_out = model.generate(input_ids = input_tokens["input_ids"],max_new_tokens=100)
gen_out
#This generates the output with heighest probability
```

```python
tokenizer.batch_decode(gen_out)
```
